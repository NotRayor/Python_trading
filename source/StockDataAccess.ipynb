```python
ticker = '005930.KS'
sam = data.get_data_yahoo(ticker, '2016-06-01')
print(sam)
```

```python
sam5 = sam['Adj Close'].rolling(window=5).mean()
print(sam5)
```

```python
sam.tail()
sam['Volume'] != 0
new_sam = sam[sam['Volume'] != 0]
```

```python
sam5 = new_sam['Adj Close'].rolling(window=5).mean()
sam5.tail(10)

# 새로운 칼럼 추가, insert 메서드!
new_sam.insert(len(new_sam.columns),"MA5", sam5)
new_sam.tail(10)
```

```python
import matplotlib.pyplot as plt

plt.plot(new_sam.index, new_sam['Adj Close'], label = "Adj Close")
```
